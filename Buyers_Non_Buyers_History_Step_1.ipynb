```python
import pandas as pd
import numpy as np
import datetime
```

```python

# Declaring an empty dataframe
buyers_histoy = pd.DataFrame()
jl = 31

for i in range(3,5,1): # Going through March and April Data
    if(i==2):
        jl = 28
    elif(i%2==0):
        jl = 31
    else:
        jl = 32
    for j in range(1,jl,1):
        a = str(i).zfill(2)
        b = str(j).zfill(2)
        dt = "2023-" + a + '-' + b
        print(dt)
        k = dt + "_clr.csv"
        l = dt + "_gbdt.csv"
        m = dt + "_shortlisting.csv"
        
#         Going through the 3 scripts individually

        clr = pd.read_csv('/home/deepaksahani_r/Desktop/ML_Prospect_Data/Combined_Mar_April/'+ k)
        clr = clr[~((clr['vis']=='#NAME?')|(clr['vis']=='#VALUE!'))].copy()
    
        gbdt = pd.read_csv('/home/deepaksahani_r/Desktop/ML_Prospect_Data/Combined_Mar_April/'+ l)
        gbdt = gbdt[~((gbdt['vis']=='#NAME?')|(gbdt['vis']=='#VALUE!'))].copy()

        sl = pd.read_csv('/home/deepaksahani_r/Desktop/ML_Prospect_Data/Combined_Mar_April/'+ m)


        temp_df = gbdt.merge(clr, how='outer', on='vis')
        dff = temp_df.merge(sl,how='outer', on='vis')
        dff.loc[((dff['ptype'] =='P1') | (dff['ptype'] =='P2')),'shortlistingintent']= 'high'
        dff.loc[((dff['ptype'] == 0)),'shortlistingintent']= 'low'
        dff.fillna(0,inplace=True)
        
        # After merging and converting SL scripts categorical values to High and Low

        dff['browsing_date'] = pd.to_datetime(dt.replace("'",""))
        dff.loc[(dff['browsing_date']==0),'browsing_date']= dff['dt_x']
        dff.loc[(dff['browsing_date']==0),'browsing_date']= dff['dt_y']
        dff.loc[(dff['browsing_date']==0),'browsing_date']= dff['dt']
        dff['browsing_date'] = pd.to_datetime(dff.browsing_date,format='%d-%m-%Y',errors='coerce')
        
        # Filling the browsing date from all the three scripts, in case Vis has only one among SL, CLR or GBDT 
        
        print(dff.shape)

        unique_buyer_vis = pd.read_csv("unique_buyer_vis.csv") # Using this to get the buyers VIS ids 
        
        fdf = dff.merge(unique_buyer_vis, how = 'left', left_on='vis', right_on='vis')
        fdf.fillna(0,inplace=True)
        fdf['trans_date'] = pd.to_datetime(fdf['trans_date']) # Getting the transaction date
        
        fdf['diff'] = (fdf['trans_date']-fdf['browsing_date']).dt.days # Getting the difference between browsing and transaction date
        res = fdf[fdf['trans_date']!='1970-01-01']
        bhdf = res[res['diff']>=1] # only more than and equal to 1 day.

        buy = bhdf[['vis','bdayscore','annivscore','spousebdayscore','engagementscore','giftingbannerscore'
               ,'giftingintent','deliveryintentscore','deliveryintent','buyingintentscore','buyingintent','trialintentscore',
               'trialintent','callintentscore','callintent','leadintentscore','leadintent','repeatvisitscore','repeatvisit','ptype',
                'shortlistingintent','browsing_date','email','trans_date','diff']].copy()
        
        buyers_histoy = pd.concat([buyers_histoy,buy],ignore_index=True) # Doing it daily and then appending in a single dataframe.
        print(buyers_histoy.shape)

    
    

    
    
    
```
